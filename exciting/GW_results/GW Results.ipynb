### Choice of Muffin Tin Sizes

Different rgkmax values are required for different species to obtain the same accuracy. Because rgkmax is set once,
according to the smallest muffin tin radius, this determines the maximum G vector. As such, to obtain the desired
rgkmax values for all other species in the calculation, their muffin tin radii must be changed proportionally to obtain
an rgkmax value associated with the same accuracy.

For a fixed input:

rgkmax_input = min(MT) * Gmax

therefore

Gmax = rgkmax_input / min(MT)

rgkmax_X = MT_X * Gmax
rgkmax_X = MT_X * rgkmax_input / min(MT)

Therefore, to obtain the desired rgkmax_X, MT_X is the free parameter:

MT_X = (rgkmax_X / rgkmax_input) * min(MT)

For Zr and O, for a given accuracy (confirm with Andris) the rgkmax values are Zr = 14-15 and O = 12.
So MT_zr = (15/12) * MT_O
