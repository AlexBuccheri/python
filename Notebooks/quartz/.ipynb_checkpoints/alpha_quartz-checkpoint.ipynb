# Lattice Parameters 


In my prior script, I demonstrated for $\alpha$-quartz how one can 
- a) Generate the lattice vectors from the lattice constants and angles
- b) Use the symmetry transformations of the space group on the basis atoms of the asymmetric cell to generate the atomic basis of the unit cell

Alternatively, as it would be cumbersome to code the symmetry operations for every space group, one can use GULP to apply the symmetry transformations.   

***Command to run GULP:***  
./gulp < quartz.in >terminal.out

## DFTB+

One does two calculations for band structures:
- a). An SCC to converge the charges. I used 24x24x24 k-points (MP grid) and an SCC tolerance of 1e-8. 
- b). An non-SCC calculation, using charges from the prior calculation as an input.
